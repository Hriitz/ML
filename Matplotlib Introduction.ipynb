# Matplolib Example workflow

```python
# importing libraries
%matplotlib inline
import matplotlib.pyplot as plt

# prepare data
x=[1,2,3,4]
y=[10,20,30,40]

# setup plot
fig, ax = plt.subplots(figsize=(5,5)) # figsize is dimension of plot (width and height)

# plot data
ax.plot(x,y);

# Customize plot
ax.set(title="Simple plot",
        xlabel="x-axis",
        ylabel="y-axis")

# Save and Show the plot
fig.savefig("../ML/images\sample-plot.png")
```

## Two options for subplots

```python
# subplot option 1
fig , ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5)); 
# plot to each different axes
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax3.set(ylabel="Prices($)")
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### pyplot vs matplotlib Object Oriented(OO) method

* when plotting something quickly, okay to use pyplot method
* when plotting something more advanced, use OO method


```python
heart_disease.head()
```

```python
over_50= heart_disease[heart_disease["age"]> 50]
over_50.head()
```

```python
len(over_50)
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method mixed wih pyplot
fig, ax= plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
#ax.set_xlim([45, 100])
```

```python
    # OO method from scratch
    fig, ax= plt.subplots(figsize=(10,6))

    # Plot data
    scatter= ax.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"]);

    # Cusotmize the plot
    ax.set(title="Heart Disease and cholesterol levels",
          xlabel="Age",
          ylabel="cholesterol")

    # Adding legends
    ax.legend(*scatter.legend_elements(), title="Target");

    # Adding horizontal line
    ax.axhline(over_50["chol"].mean(),
              linestyle="--")
```

```python
over_50.head()
```

```python
# Subplots of chol and thalach
fig, (ax0, ax1)= plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex= True)
# Add data to axis 0
scatter= ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize axis 0
ax0.set(title="Heart disease and cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Adding legends
ax0.legend(*scatter.legend_elements(), title="target")

# Adding mean cholesterol line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")

# Add data to axis 1
scatter= ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])
# Customize axis 1
ax1.set(title="Heart disease and max heart rate levels",
       xlabel="Age",
       ylabel="max heart rate")

# Adding legends to axis 1
ax1.legend(*scatter.legend_elements(), title="target")

# Adding mean heart rate line
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Adding title to figure
fig.suptitle("Heart Disease Analysis", fontsize="16", fontweight="bold" );
```
